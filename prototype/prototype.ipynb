# ner result

```python
ner_token_dir = './output_ner_tokenized.xlsx'
all_text = pd.read_excel(ner_token_dir).values

ner_label_dir = './output_ner_pred.xlsx'
all_label = pd.read_excel(ner_label_dir).iloc[:,1:].values
```

```python
def extract_ner(text, label):
    pred_tags = [int_to_label[i] for i in label]
    pred_texts = text
    result = [[text, tag] for tag,text in zip(pred_tags, pred_texts) if tag!='PAD' and tag!='O']
    
    return result
```

```python
def recover_tokenize(sentence):
    temp = []
    for i in sentence :
        if str(i)=='nan' :
            continue
        if str(i).find('##')==-1:
            temp.append(' '+str(i))
        else :
            temp.append(str(i).replace('##','').replace('에','').replace('는','').replace(')',''))
    return ''.join(temp).strip()
```

```python
def tag_select(extract, tag):
    if tag == '회사':
        return [i[0] for i in extract if i[1]=='B-cor' or i[1]=='I-cor']
    elif tag == '이자율':
        return [i[0] for i in extract if i[1]=='B-ir' or i[1]=='I-ir']
    elif tag == '수익률':
        return [i[0] for i in extract if i[1]=='B-er' or i[1]=='I-er']
    elif tag == '배정':
        return [i[0] for i in extract if i[1]=='B-s' or i[1]=='I-s']
    elif tag == '잔액':
        return [i[0] for i in extract if i[1]=='B-a' or i[1]=='I-a']
    else :
        return None
```

```python
def input_query(first_query, second_query):
    first_query_idx = np.array([idx for idx, i in enumerate(clas_res['predict_label'].values) if i==first_query])
    if len(first_query_idx)==0 :
        print('찾으시는 조항 정보가 탐색되지 않습니다')
        return None
    
    all_extract = [extract_ner(text, label) for text, label in zip(all_text[first_query_idx], all_label[first_query_idx])]
    select_extract = [tag_select(i, second_query) for i in all_extract]
    if select_extract[0]==None:
        print('찾으시는 추출 정보가 탐색되지 않습니다')
        return None
    
    result = pd.concat([clas_res.iloc[first_query_idx].reset_index(), pd.DataFrame([recover_tokenize(i) for i in select_extract]).reset_index()], axis=1)
    result = result.drop(['index'],axis=1)
    result.columns=['계약서 조항 내용', '클래스', '결과']
#     result = result[result['결과']!='']
    dfStyler = result.style.set_properties(**{'text-align': 'left'})
    result = dfStyler.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
    print('찾으시는 결과는 다음과 같습니다')
    return result
```
